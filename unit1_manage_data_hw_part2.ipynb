When we want to make changes to the data, we will make a function that takes a single row and changes the element based on certain conditions. To do this we will use our first conditional statement using `if` else logic. It's okay if you don't fully understand the code below, it will make more sense as we go along. (Time can be complicated)

```python
def mil_time(row):
  if 1 <= row['time_sugar_hr'] <= 11:
    if row['time_sugar_ampm'] == 1:
      return row['time_sugar_hr']
    elif row['time_sugar_ampm'] == 2:
      return row['time_sugar_hr'] + 12
    else:
      return np.nan
  elif row['time_sugar_hr'] == 12:
    if row['time_sugar_ampm'] == 1:
      return 0
    elif row['time_sugar_ampm'] == 2:
      return 12
    else:
      return np.nan
  else:
    return np.nan
```

Now let's apply this to each row of our data set and save our result into a new variable. 

```python
myData['time_sugar_mt'] = myData.apply(mil_time,axis=1)
```
